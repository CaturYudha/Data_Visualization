```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data bank.csv")
plt.scatter(data['duration'], data['job'])
plt.title("scatter plot")
plt.savefig("output1.jpg")
plt.xlabel('duration')
plt.ylabel('job')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data bank.csv")
plt.plot(data['balance'])
plt.plot(data['duration'])
plt.title("Line Chart")
plt.savefig("output2.jpg")
plt.xlabel("balance")
plt.ylabel("duration")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data bank.csv")
plt.bar(data['job'], data['duration'])
plt.title("Bar Chart")
plt.savefig("output3.jpg")
plt.xlabel("job")
plt.ylabel("duration")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data bank.csv")
plt.hist(data['balance'])
plt.title("historygram Chart")
plt.savefig("output3.jpg")

plt.show()
```
